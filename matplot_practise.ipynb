Letâ€™s track the daily sales of a coffee shop for a week.

Example:

* X-axis: Days of the week.
* Y-axis: Number of cups of coffee sold.

Steps:

```python
import matplotlib.pyplot as plt
days= ['mon','tues','wed','thurs','fri','sat','sun']
coffee_sales= [50,60,70,65,80,90,100]

plt.plot(days,coffee_sales)
plt.show()
```

```python
import matplotlib.pyplot as plt
days= ['mon','tues','wed','thurs','fri','sat','sun']
coffee_sales= [50,60,70,65,80,90,100]
tea_sales= [30, 40, 35, 45, 50, 60, 70]

plt.plot(days,coffee_sales,tea_sales)
plt.show()
```

```python
plt.plot(days,coffee_sales,linestyle='-.',color= 'red',marker='o')
plt.plot(days,tea_sales,linestyle='--',color= 'blue',marker='s')
plt.xlabel('Days of the week')
plt.ylabel('Coffee and tea sales')
plt.title('coffee sales over a week')
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot(days,coffee_sales,linestyle='-.',color= 'red',marker='o',label= 'tea')
plt.plot(days,tea_sales,linestyle='--',color= 'blue',marker='s',label='coffee')
plt.xlabel('Days of the week')
plt.ylabel('Coffee and tea sales')
plt.title('coffee sales over a week')
plt.legend()
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot(days,coffee_sales,linestyle='-.',color= 'red',marker='o',label= 'tea')
plt.plot(days,tea_sales,linestyle='--',color= 'blue',marker='s',label='coffee')

for i,j in zip(days,coffee_sales):
   plt.text(i,j+1,str(j),ha='center',va='bottom',fontsize=10,color= 'brown')

for i,j in zip(days,tea_sales):
   plt.text(i,j+1,str(j),ha='center',va='bottom',fontsize=10,color='black')

plt.xlabel('Days of the week')
plt.ylabel('Coffee and tea sales')
plt.title('coffee sales over a week')
plt.legend()
plt.grid()
plt.show()
```

Creating Trend Lines: Revenue Growth Over a Year

Real-World Example:

Youâ€™re analyzing monthly revenue for your online store over a year and want to predict future performance.

Steps to Add a Trend Line:

1.Data:

X-axis: Months of the year.
Y-axis: Monthly revenue (e.g., [5000, 5200, 5500, 5700, 6000, 6200, 6400, 6600, 6800, 7000, 7200, 7500]).
2.Use a Trend Line:

Fit a line through the data points to show the overall revenue growth trend.
3.Interpretation:

The trend line shows that revenue is steadily increasing and can help you predict next monthâ€™s earnings.

```python
import numpy as np
import matplotlib.pyplot  as plt
Months=['Jan','feb','Mar','Apr','may','june','jul','Aug','Sep','oct','Nov','dec']
revenue=[5000,5200,5500,5700,6000,6200,6400,6600,6800,7000,7200,7500]

# Use numerical representation for months for polyfit
x_months = np.arange(len(Months))

z= np.polyfit(x_months,revenue,1)
p= np.poly1d(z)
trendline= p(x_months)

plt.plot(Months,revenue,linestyle= '--',color='green',marker='o',label= 'revenue')
plt.plot(Months,trendline,color= 'skyblue',label='trendline')

for i, j in zip(Months,revenue):
  plt.text(i,j+1,str(j),ha= 'center',va='bottom',color='black',fontsize=8)
plt.xlabel('Months of the year')
plt.ylabel('Revenue per Month')
plt.title('Revenue Growth Over a Year with Trend Line')
plt.legend()
plt.grid()
plt.show()
```
